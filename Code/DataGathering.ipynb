```python
import pandas as pd
from pathlib import Path
import calendar
```

```python
# Read in the CSV file as a Pandas DataFrame
jan2023 = pd.read_csv(Path("C:/Users/fishm/Documents/UT Bootcamp/HW13/data/jan2023.csv"))
may2023 = pd.read_csv(Path("C:/Users/fishm/Documents/UT Bootcamp/HW13/data/may2023.csv"))
sep2023 = pd.read_csv(Path("C:/Users/fishm/Documents/UT Bootcamp/HW13/data/sep2023.csv"))
```

```python
# Combine the DataFrames vertically
combined_df = pd.concat([jan2023, may2023, sep2023], ignore_index=True)
combined_df
```

```python
# Assuming your date column is named "Date" in the combined DataFrame
combined_df['Month'] = pd.to_datetime(combined_df['started_at']).dt.month
combined_df
```

```python
# Define a function to convert month numbers to month names
def month_number_to_name(month_number):
    return calendar.month_name[month_number]

# Apply the function to create a new column with month names
combined_df['MonthName'] = combined_df['Month'].apply(month_number_to_name)
combined_df
```

```python
# Export the DataFrame to a CSV file
combined_df.to_csv("combined_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# Data
segments = ['Fans Attending 10+ Games', 'Fans Attending 10- Games']
allocation = [350, 150]  # Number of messages allocated to each segment

# Create a pie chart
plt.figure(figsize=(8, 8))
colors = ['black', 'red']
patches, texts, autotexts = plt.pie(allocation, labels=segments, autopct='%1.1f%%', startangle=5, colors=colors, textprops={'color': 'white'})

# Set the text color of percentage labels to white
for text in autotexts:
    text.set_color('white')

plt.title('Messaging Allocation Breakdown')

# Display the chart
plt.axis('equal')

# Export to an image file (e.g., PNG)
plt.savefig('AllocationBreakdown.png')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
segments = ['Income $100,000 - 249,999', 'Income $250,000+', 'Ages 45-54', 'Ages 65+', 'Higher Zip Codes']
allocation = [20, 30, 20, 25, 5]  # Percentage allocation for each segment

# Create a pie chart
fig, ax = plt.subplots(figsize=(8, 8))  # Create figure and axis objects
ax.pie(allocation, labels=segments, autopct='%1.1f%%', startangle=0, colors=['yellow', 'green', 'orange', 'purple', 'lightblue'])
ax.set_title('Messaging Allocation for Fans Attending 10 or Games (350)')

# Display the chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
fig.tight_layout()

# Export to an image file (e.g., PNG)
plt.savefig('10plus.png')

plt.show()
```
